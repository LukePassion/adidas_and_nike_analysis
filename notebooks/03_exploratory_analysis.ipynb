```python
#Carregando arquivos

df = spark.table("workspace.default.adidas_nike_balanced")
```

```python
display(df)
```

```python
#Comparação de preço médio por marca
#Qual marca é mais cara?

#Criando coluna para apresentar os preços corretamente

from pyspark.sql.functions import col

df = df.withColumn("sale_price_real", col("sale_price") / 100)

#Analisando média de preços
from pyspark.sql.functions import avg

df.groupBy("brand_clean") \
.agg(avg("sale_price_real").alias("avg_price")) \
.display()
```

```python
df.select("sale_price_real").show(10)
```

```python
#Comparação de desconto médio
#Quem oferece mais desconto?

df.groupBy("brand_clean") \
  .agg(avg("discount_percentage").alias("avg_discount")) \
  .display()
```

```python
#Popularidade média
#Qual marca gera mais engajamento?

df.groupBy("brand_clean") \
  .agg(avg("reviews").alias("avg_reviews")) \
  .display()
```

```python
#Avaliação média (rating)
#Qual marca é melhor avaliada?

df.groupBy("brand_clean") \
  .agg(avg("rating").alias("avg_rating")) \
  .display()
```

```python
#Distribuição de categorias de preço
#Cada marca atua mais em que faixa de preço?

from pyspark.sql.functions import avg, min, max

df.groupBy("brand_clean") \
  .agg(
      avg("sale_price").alias("avg_price"),
      min("sale_price").alias("min_price"),
      max("sale_price").alias("max_price")
  ).display()
```

```python
#Relação preço vs avaliação
#Produtos mais caros têm melhores avaliações?

df.select("sale_price", "rating", "brand_clean")
```

```python
#Produtos mais populares
#Quais produtos dominam atenção?

df.orderBy("reviews", ascending=False) \
  .select("product_name", "brand_clean", "reviews") \
  .display(10)
```

```python
df.write.mode("overwrite").saveAsTable("adidas_nike_clean")
```

##Conclusão da Análise exploratória

A análise evidencia duas estratégias bem distintas entre as marcas:
	
  •	NIKE: se posiciona como uma marca premium moderada, com preços mais elevados e produtos de destaque que concentram grande atenção dos consumidores.
	
  •	ADIDAS: adota uma estratégia mais orientada a volume, com preços mais acessíveis, maiores descontos e maior engajamento dos clientes.

Como resultado, a Adidas demonstra um relacionamento mais amplo e consistente com o público, enquanto a Nike constrói valor através de exclusividade e posicionamento de marca.
