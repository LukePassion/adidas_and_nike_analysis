```python
#Carregando arquivos

df = spark.table("workspace.default.adidas_nike_clean")
```

```python
#Converter para Pandas

pdf = df.toPandas()
```

```python
#1. Preço médio por marca

import matplotlib.pyplot as plt

price_mean = pdf.groupby("brand_clean")["sale_price_real"].mean()

price_mean.plot(kind="bar")
plt.title("Preço Médio por Marca")
plt.ylabel("Preço (R$)")
plt.xlabel("Marca")
plt.show()
```

##1. Preço médio por marca:
##Nike é mais cara que Adidas

```python
#2. Desconto médio por marca

discount_mean = pdf.groupby("brand_clean")["discount_percentage"].mean()

discount_mean.plot(kind="bar")
plt.title("Desconto Médio por Marca")
plt.ylabel("% de Desconto")
plt.show()
```

##2. Desconto por marca:
##Adidas oferece mais desconto

```python
#3.Popularidade média por marca

reviews_mean = pdf.groupby("brand_clean")["reviews"].mean()

reviews_mean.plot(kind="bar")
plt.title("Popularidade Média (Reviews)")
plt.ylabel("Número de Reviews")
plt.show()
```

##Popularidade média por marca
##Adidas gera mais engajamento

```python
#4.Avaliação média

rating_mean = pdf.groupby("brand_clean")["rating"].mean()

rating_mean.plot(kind="bar")
plt.title("Avaliação Média por Marca")
plt.ylabel("Rating")
plt.show()
```

##Avaliação média
##Adidas tem melhor avaliação média

```python
#5. Distribuição de preços

import seaborn as sns

sns.boxplot(x="brand_clean", y="sale_price_real", data=pdf)
plt.title("Distribuição de Preços por Marca")
plt.show()
```

##Distribuição de preços
##Mostra dispersão + posicionamento real

```python
#6. Relação preço vs avaliação

sns.scatterplot(x="sale_price_real", y="rating", hue="brand_clean", data=pdf)
plt.title("Preço vs Avaliação")
plt.show()
```

##Relação preço vs avaliação
##Produtos mais caros nem sempre são os melhores avaliados

```python
#7.Top produtos mais populares

top_products = pdf.sort_values(by="reviews", ascending=False).head(10)

plt.barh(top_products["product_name"], top_products["reviews"])
plt.title("Top 10 Produtos Mais Populares")
plt.xlabel("Reviews")
plt.show()
```

##Top produtos mais populares
##Destaque para produtos Nike + dominío Adidas no volume

##Isights Visuais
•	A Nike apresenta preços mais elevados, confirmando posicionamento premium
	
  •	A Adidas se destaca em descontos e volume de produtos
	
  •	A Adidas possui maior engajamento e melhor avaliação média
	
  •	A distribuição de preços mostra maior acessibilidade da Adidas
	
  •	Não há relação direta entre preço e avaliação
	
  •	Produtos da Nike lideram em destaque individual, enquanto a Adidas domina em escala
