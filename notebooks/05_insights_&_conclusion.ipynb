##1.Objetivo do projeto
O objetivo deste projeto foi analisar e comparar os produtos das marcas Adidas e Nike, buscando entender diferenças em preço, avaliação, popularidade e estratégia de mercado.


##2.Posicionamento de preço
	
  •	Nike apresenta preços médios mais elevados
	
  •	Adidas atua em uma faixa mais acessível

##Interpretação:

Nike se posiciona como marca premium, enquanto Adidas adota estratégia mais acessível

##3.Estratégia de descontos
	
  •	Adidas oferece descontos significativamente maiores
	
  •	Nike mantém descontos mais baixos

##Interpretação:

Adidas utiliza preço e promoção como vantagem competitiva

##4.Engajamento (Popularidade)
	
  •	Adidas possui maior número médio de reviews
	
  •	Forte presença no ranking de produtos mais populares

##Interpretação:

Adidas gera maior volume de interação com clientes

##5.Avaliação dos clientes
	
  •	Adidas possui rating médio superior
	
  •	Nike apresenta avaliações mais baixas

##interpretação:

A experiência do cliente tende a ser melhor com produtos Adidas

##Insight

_Enquanto a Nike constrói valor através de posicionamento premium e marca forte, a Adidas conquista o cliente através de acessibilidade, volume e maior percepção de valor._

##Resposta à pergunta principal

Qual marca apresenta melhor relação entre preço, avaliação e engajamento?

**Resposta:**

A Adidas apresenta melhor equilíbrio geral entre preço, avaliação e engajamento, enquanto a Nike se destaca pelo posicionamento premium, porém com menor percepção de valor pelos clientes.

##6.Conclusão Final:

Este projeto demonstrou como técnicas de análise de dados podem revelar padrões importantes de comportamento do consumidor e posicionamento de mercado, auxiliando na tomada de decisões estratégicas baseadas em dados.
